```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

# loading data
df = pd.read_csv("/Users/maxwellpearse/Library/Mobile Documents/com~apple~CloudDocs/Life/b.Business, Money, Education,  Employment/Python/data_jobs.csv")

# cleaning data
df["job_posted_data"] = pd.to_datetime(df.job_posted_date)
```

```python
df_usa = df[df["job_country"]== "United States"].copy() # Only countries from the united states
```

```python
df_usa["job_posted_monthly"] = df_usa["job_posted_data"].dt.strftime("%B") #dt allows you to apply datetime functions to column

df_usa
```

```python
df_usa_pivot = df_usa.pivot_table(index="job_posted_monthly", columns="job_title_short", aggfunc="size")
```

```python
df_usa_pivot.reset_index(inplace=True)
df_usa_pivot["month no"] = pd.to_datetime(df_usa_pivot["job_posted_monthly"],format="%B").dt.month

df_usa_pivot["month no"]
```

```python
df_usa_pivot.sort_values("month no", inplace=True)
df_usa_pivot.set_index("job_posted_monthly", inplace=True)
df_usa_pivot.drop(columns="month no",inplace=True)
```

```python
df_usa_pivot
```

```python
df_usa_pivot.plot(kind="line")
top3 = df_usa["job_title_short"].value_counts().head(3)
top3 = top3.index.to_list()
```

```python
df_usa_pivot[top3].plot(kind="line")
plt.title("Monthly Job postings for Top Data Jobs in the US")
plt.ylabel("Job Count")
plt.xlabel("2023")
plt.legend() # removes the title of the legend box
plt.show()
```

```python
# import libraries
import ast
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

# loading data
df = pd.read_csv("/Users/maxwellpearse/Library/Mobile Documents/com~apple~CloudDocs/Life/b.Business, Money, Education,  Employment/Python/data_jobs.csv")

# cleaning data
df["job_posted_date"] = pd.to_datetime(df.job_posted_date)
df["job_skills"] = df["job_skills"].apply(lambda skill_list: ast.literal_eval(skill_list) if pd.notna(skill_list) else skill_list)
```

```python
df_DA = df[df["job_title_short"] == "Data Analyst"].copy()
```

```python
df_DA["job_posted_month_no"] = df_DA["job_posted_date"].dt.month

df_DA
```

```python
df_DA_explode = df_DA.explode("job_skills")

df_DA_pivot = df_DA_explode.pivot_table(index="job_posted_month_no", columns="job_skills", aggfunc="size", fill_value=0)
```

```python
df_DA_pivot.loc["Total"] = df_DA_pivot.sum() #note this makes a new row!!!!!!!
df_DA_pivot = df_DA_pivot[df_DA_pivot.loc["Total"].sort_values(ascending=False).index] # sorting the columns by inside part
df_DA_pivot = df_DA_pivot.drop("Total")
df_DA_pivot
```

```python
df_DA_pivot.iloc[:, :5].plot(kind="line")
plt.title("Top 5 Skills for Data Analyst per Month")
plt.xlabel("")
plt.ylabel("Count")
plt.show()
```

## Matplotlib Advanced customisation

```python
df_DA_pivot.iloc[:, :5].plot(
    kind="line",
    linewidth=4,
    linestyle="-.",  # options: '-','--','-.',':'
    colormap="viridis", # about four more on cheat sheet
    marker="o",
    markersize=5,
    figsize=(7,3.5),
    )

plt.title("Top 5 Skills for Data Analyst per Month")
plt.xlabel("")
plt.ylabel("Count")
plt.legend()
plt.show()
```
